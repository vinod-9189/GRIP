```python
# IMPORTING REQUIRED LIBRARIES.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
pd.set_option('display.max_rows', 10)
```

```python
# IMPORT DATASET.

DATA = pd.read_csv("DATA.csv")
DATA.describe()
```

```python
# CHECK FOR OUTLIERS (INDIVIDUAL COLUMNS)

fig, axs = plt.subplots(figsize=(20, 5),ncols=4)
sns.boxplot(x=DATA["Hours"], ax=axs[0])
sns.boxplot(x=DATA["Scores"], ax=axs[1])

# CHECK FOR NORMALITY

sns.kdeplot(data=DATA, x="Hours", ax=axs[2])
sns.kdeplot(data=DATA, x="Scores", ax=axs[3])
```

```python
# REGRESSION POLT

plt.figure(figsize=(8,6))
sns.regplot(data=DATA, x="Hours", y="Scores")
plt.title("REGRESSION PLOT")
```

```python
# DATA SPLIT INTO (TRAIN & TEST)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(DATA["Scores"], DATA["Hours"], test_size=0.33, random_state=42)

X_TRAIN = np.array(X_TRAIN).reshape(-1,1)
X_TEST = np.array(X_TEST).reshape(-1,1)

# MODEL CREATION

MODEL = LinearRegression()
MODEL.fit(X_TRAIN,Y_TRAIN)
print("R^2 ON TRAINING DATA",MODEL.score(X_TRAIN,Y_TRAIN))
```

```python
# TESTING THE MODEL

print("R^2 ON TEST DATA",MODEL.score(X_TEST,Y_TEST))
```
